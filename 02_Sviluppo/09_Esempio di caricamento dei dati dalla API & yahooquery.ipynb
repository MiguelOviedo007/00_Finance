## Misure di valutazione

Note:
Trailing 12 months (TTM) è il termine per i dati degli ultimi 12 mesi consecutivi utilizzati per riportare i dati finanziari.
I 12 mesi finali di una società rappresentano la sua performance finanziaria per un periodo di 12 mesi; in genere non rappresenta un periodo di fine anno fiscale.

```python
aapl = Ticker('msft')
misure_valutazione=aapl.valuation_measures.reset_index().set_index('asOfDate')
misure_valutazione
```

### MarketCap più recente

```python
misure_valutazione.loc[(misure_valutazione['periodType']=='TTM') & (misure_valutazione['MarketCap'].notna())].max()['MarketCap']
#Prendere il date più recente
```

### PeRatio

```python
misure_valutazione.loc[(misure_valutazione['periodType']=='TTM') & (misure_valutazione['PeRatio'].notna())].max()['PeRatio']
```

### PeRatio fordward

```python
misure_valutazione.loc[(misure_valutazione['periodType']=='TTM') & (misure_valutazione['ForwardPeRatio'].notna())].max()['ForwardPeRatio']
```
