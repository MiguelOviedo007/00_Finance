# Frontiera Efficiente

```python
rportafoglio = [] #Rendimenti
stdportafoglio = []#Dev standard
lista_pesi = []
```

 * Genero 5000 combinazioni di pesi casuali
 * Calcolo il rendimento e la dev standard per ogni combinazioni
 * Salvo i risultati nelle liste

```python
pesi = np.random.random(2)
```

```python
for x in range(5000):#5000 combinazioni di pesi...
    pesi = np.random.random(2)
    pesi /= np.sum(pesi)#serve per fare in modo che i pesi sommino 1
    lista_pesi.append(pesi)
    rportafoglio.append(np.dot(due_stocks.mean() * 12, pesi))
    stdportafoglio.append(np.sqrt(np.dot(pesi.T, np.dot(due_stocks.cov() * 12, pesi ))))
```

```python
np.array(lista_pesi)
```

```python
pesiportafoglio = np.array(lista_pesi)
rportafoglio = np.array(rportafoglio)
stdportafoglio = np.array(stdportafoglio)
```

```python
pesiportafoglio[:, 0]
```

```python
portafoglio = pd.DataFrame({"peso_1" : pesiportafoglio[:, 0], "peso_2" : pesiportafoglio[:, 1] ,'Rendimenti':rportafoglio , 'Volatilita':stdportafoglio})
```

```python
portafoglio
```

```python
import matplotlib.pyplot as plt
```

```python
portafoglio.plot(x = 'Volatilita', y='Rendimenti', kind='scatter', title='Frontiera Efficiente');
```
